**Oppgave 1**
Vi ser i første omgang på en numerisk integrasjonsformel på intervallet $[0,1]$ av formen

$$
    Q[f] = s\left(f(1)-f(0)\right) + b f(\alpha) 
$$

der $s$, $b$ og $\alpha$ er parametre som kan velges.


<div class="alert alert-block alert-success">
**Oppgave 1a.**
Finn verdier for $s$, $b$ og $\alpha$ som gjør at formelen får høyest mulig presisjonsgrad.
</div>


**Kontrollspørsmål 1:**
Hva er den største mulige presisjonsgrad?


**SVAR:**

*Merk: Omdøper* $b$ *til* $\beta$ *i kun denne oppgaven, for enkelhets skyld. Altså er* $ Q[f] = s\left(f(1)-f(0)\right) + \beta f(\alpha) $.

Vi lar $f(x)=ax^3+bx^2+cx+d$, og utleder hvilke krav dette stiller til $(s, \alpha, \beta)$ fra den laveste graden og opp, og ser hvor langt vi kommer før vi må sette koeffisientene i $f$ til 0.

Vi bruker at for at vi skal ha presisjonsgrad av samme grad som polynomet $f$ må $Q[f] = \int_0^1fdx$, så dermed må

$$
\begin{split}
Q[f] &= s\left(f(1)-f(0)\right) + \beta f(\alpha)\; &= s((a+b+c+d) - d) + \beta(a\alpha^3 + b\alpha^2 + c\alpha + d)  \\
&= \int_0^1f\ dx &= \frac{a}4 + \frac{b}3 + \frac{c}2 + d\\
\implies \; & d(\beta-1)=0 &\implies \beta=1 \\
\implies\;& c(s+\alpha-\frac12) = 0 &\implies s = \frac12 - \alpha \\
\implies\;& b\left(\frac12-\alpha+\alpha^2-\frac13\right)=b\left(\alpha^2-\alpha+\frac16\right)=0\;
&\implies \alpha = \frac12 \left(1\pm \frac1{\sqrt3}\right) \implies s=\mp \frac1{2\sqrt3}\\
\implies\; & a\left[\mp \frac1{2\sqrt3} + \frac1{2^3} \left(1\pm \frac1{\sqrt3}\right)^3 - \frac14\right]=0 &\implies a\left(\mp\frac{\sqrt3}{36}\right)=0\implies a=0\\
\end{split}
$$
Så presisjonsgraden kan ikke være 3, men med $(s, \alpha, \beta) = \left(\mp \frac1{2\sqrt3}, \frac12 \left(1\pm \frac1{\sqrt3}\right), 1\right)$ har vi **presisjonsgrad 2**.

**Kontrollspørsmål 2:** Det fins to løsninger for $\alpha$, hva er verdien av den minste av disse?

$\alpha_- = \frac12 \left(1 - \frac1{\sqrt3}\right) \simeq 0.21$



Vi modifiserer nå formelen ved å transformere den til intervallet $[x_k,x_k+h]$.
Et lineært variabelskifte, $x=x_k+th$, gir oss

$$
  \int_{x_k}^{x_k+h}f(x)\,dx = h\int_0^1 f(x_k+th)\,dt 
$$

som gir den modifiserte formelen

$$
     Q_{[x_k,x_k+h]}[f] = h\big(s(f(x_k+h)-f(x_k)) + b f(x_k+\alpha h)\big)
$$

Neste skritt er å lage en sammensatt formel; definer $x_k=a+kh$ der $h=\frac{b-a}{n}$ og $n$ er antall intervaller. Vi kan nå lage den sammensatte formelen

$$
  \bar{Q}_n[f] = \sum_{k=0}^{n-1} Q_{[x_k,x_k+h]}[f]
$$

</div>

Med $(s, \alpha, b) = \left(\frac1{2\sqrt3}, \frac12 \left(1 - \frac1{\sqrt3}\right), 1\right)$ har vi at
$$
\bar{Q}_n[f] = h \sum_{k=0}^{n-1} \left\{\frac1{2\sqrt3}(f(x_k+h)-f(x_k)) + f\left(x_k+\frac12 \left(1 -1/\sqrt3\right) h\right)\right\}
$$

**Kontrollspørsmål 3:**
Hvor mange funksjonsevalueringer (beregninger av $f(x)$) trenger man med $n$ intervaller?

**SVAR:**
Vi sjekker hver "kantverdi" på hvert intervall, og bruker dem i to intervaller. Dette gir $n+1$ sjekker. I tillegg sjekker vi én verdi på hvert intervall. Dette gir $n$ sjekker. **Totalt antall sjekker er $2n+1$.** Jeg er ikke enig i at det er $n+2$ uten noen overbevisende argument for det.

**Oppgave 1c.**
Implementer metoden i Python. Anvend den på integralet

$$
I =   \int_{-2}^2 \frac{\sin x}{x}\,dx
$$

med $n=4,8,16,32,64,128$ og sammenlign svaret med sammensatt trapesmetode med samme antall intervaller.
Husk unntaksbehandling av $x=0$ (spesielt relevant for trapes).

Fyll inn koden i cellen nedenfor

```python
import numpy as np
import warnings

def f(x):
    '''Definer funksjonen f(x) her'''
    if type(x) is np.ndarray:
        with warnings.catch_warnings():
            warnings.simplefilter("ignore")
            ret = np.sin(x)/x
        ret[x==0] = 1
        return ret
    
    else:
        return np.sin(x)/x if x != 0 else 1
    

def teleformula(f,a,b,n,c):
    
    '''Beregner approksimasjon til I basert på formelen utledet i denne oppgaven
    f: integranden
    a,b: Intervallgrenser
    n: Antall intervaller
    c: skal være 1 eller -1 og velger hvilken av de to metodene med presisjonsgrad 2 
        som skal brukes. Negativ c gir minste alpha
    '''
    assert c in (-1, 1) and a <= b and type(n) in {int, np.intc} and n > 0, f"c: {c}, (a, b): {a, b}, type(n): {type(n)}."
    
    h = (b-a) / n
    alpha = 1/2 + c / (2*np.sqrt(3))
    s = -c / (2*np.sqrt(3))
    
    x = np.arange(n)*h + a
    
    addends = s*(f(x+h) - f(x)) + f(x + h*alpha)
    
    return h*np.sum(addends)


def trapesformula(f,a,b,n):
    '''For sammenligning implementer sammensatt trapes her'''
    x = np.linspace(a, b, n+1)
    x_weight = np.ones_like(x)
    x_weight[0] = x_weight[-1] = 1/2
    x_weight *= (b-a) / np.sum(x_weight)
    
    return np.sum(f(x)*x_weight)


# n_arr = [4,8,16,32,64,128,256]
n_arr = 2 ** np.arange(2, 14)
I = 3.2108259536053896972  # Har funnet 'eksakt svar' på integralet for sammenligning


# ------------------------------------------------------------------------------------------
# Print simuleringsresultater:
# ------------------------------------------------------------------------------------------

print("Den nye metoden:")
for n in n_arr:
    tel = teleformula(f, -2, 2, n, 1)
    print(f"{n:3}: \t{tel:<16}\tError: {I-tel:.4G}")
print(f"Eksakt:\t{I}")

print("\nTrapesmetoden:")
for n in n_arr:
    tra = trapesformula(f, -2, 2, n)
    print(f"{n:3}: \t{tra:0<16}\tError: {I-tra:.4G}")
print(f"Eksakt:\t{I}")
```

Vi ser altså her at feilen i den nye formelen oppfører seg som $C h^4$, dette skyldes at integranden er symmetrisk.

**Kontrollspørsmål 4.** Fra eksperimentene kan man estimere at feilen i den nye formelen på dette problemet oppfører seg som $|I-\bar{Q}_n[f]|\approx C n^q$ for et heltall $q$. Bestem $q$.

**Svar:**
Vi vet at feilen går som $C h^4 = C \left(\frac{b-a}n\right)^4=C(b-a)\cdot n^{-4} = C'n^{-4} \implies q=-4$ i dette tilfellet med symmetrisk integrand.
